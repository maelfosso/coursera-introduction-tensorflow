```python
import tensorflow as tf
import numpy as np
```

# The 'Hello World' of neural networks

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)
```

```python
xs = np.array([-1.0, 0.0, 1., 2., 3., 4.], dtype=np.float)
ys = np.array([-3.0, -1., 1., 3., 5., 7.], dtype=np.float)
```

```python
model.fit(xs, ys, epochs=500)
```

```python
model.predict([10.0])
```

```python
model.predict([0])
```

```python
model.predict([1])
```
